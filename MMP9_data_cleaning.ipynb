# Data set aquisition and data cleaning from ChEMBL (MMP9, CHEMBL321)

requires ChEMBL API installation (chembl_webresource_client)

### Data aquisition

```python
import pandas as pd
from chembl_webresource_client.new_client import new_client

pd.set_option('display.max_columns', None)
```

```python
target = new_client.target
target_query = target.search('Matrix metalloproteinase 9')
targets = pd.DataFrame.from_dict(target_query)
targets
```

```python
selected_target = targets.target_chembl_id[0]

activity = new_client.activity
res = activity.filter(target_chembl_id=selected_target).filter(standard_type='IC50')
```

```python
df = pd.DataFrame.from_dict(res)
df.head()
```

```python
df.to_csv('data/sm_library_full.csv', index=False)
```

### Data cleaning and activity classification

```python
df2 = df[df.standard_value.notna()]
df2 = df2[df2.canonical_smiles.notna()]
df2.drop_duplicates(['canonical_smiles'], inplace=True)
df2.shape
```

```python
sele = ['molecule_chembl_id', 'canonical_smiles', 'standard_value']
df3 = df2[sele]
df3.to_csv('data/MMP9_sm_activity.csv', index=False)
```

```python
bioactivity_level = []
for i in df3['standard_value']:
    if float(i) >= 10000:
        bioactivity_level.append('inactive')
    elif float(i) <= 1000:
        bioactivity_level.append('active')
    elif float(i) == 0:
        bioactivity_level.append('none')
    else:
        bioactivity_level.append('intermediate')
```

```python
df3['activity_class'] = bioactivity_level
```

```python
df3.to_csv('data/MMP9_sm_activity_proc1.csv', index=False)
```
